Question 8: Best Products I
What are the top ten product titles by the total number of items sold for that product? Display in descending order. Store in variable top_num_sales.

```python
top_num_sales = df_cleaned.groupby('prod_title')['trans_quantity'].sum().sort_values(ascending=False)
top_num_sales = top_num_sales.head(10).index.tolist()
```

Question 9: Best Products II

```python
top_tot_sales = df_cleaned.groupby('prod_title')['total_sales'].sum().sort_values(ascending=False)
top_tot_sales = top_tot_sales.head(10).index.tolist()
```

Question 10: Bonus
What is the proportion of returning customers? Store as variable prop_returning

```python
# Count the number of unique customers
unique_customers = df_cleaned['cust_id'].nunique()

# Count the number of unique transactions
unique_transactions = df_cleaned['trans_id'].nunique()

# Count the number of returning customers (customers with more than one unique transaction)
returning_customers = (df_cleaned.groupby('cust_id')['trans_id'].nunique() > 1).sum()

# Calculate the proportion of returning customers
prop_returning = returning_customers / unique_customers

# Display the result
print("Proportion of returning customers:", prop_returning)
```
